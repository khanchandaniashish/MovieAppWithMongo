```python
import pandas as pd

reviews = '/Users/ashish/Documents/Semester 2/NoSQL/Team Project/dataset/output.json'
movies = '/Users/ashish/Documents/Semester 2/NoSQL/Team Project/dataset/metadata.json'

#load into dataframes
df1 = pd.read_json('/Users/ashish/Documents/Semester 2/NoSQL/Team Project/dataset/metadata.json')
df2 = pd.read_json(reviews)
```

```python
#merge and preserve the item_id field
merged_df = pd.merge(df1, df2, on='item_id', how='outer')
```

```python
merged_df.iloc[:10]
```

```python
#aggregate the reviews into array
grouped_df = merged_df.groupby(['item_id','title','directedBy','starring','avgRating','imdbId']).agg({'txt': lambda x: list(x)}).reset_index()
```

```python
output = grouped_df.to_dict(orient='records')
```
